````python
model.fit(x_train, y_train, epochs=100, verbose=4)
````

# 4.4 Modeling linear and non linear relationships

Generate two new subsets filtering these characteristics:
- D1D1: X='RM'X='RM', y='MEDV'y='MEDV'
- D2D2: X='LSTAT'X='LSTAT', y='MEDV'y='MEDV'

````python
col = ['RM','MEDV']
D1=data2[col]
print(D1.head())
````

````python
col = ['LSTAT','MEDV']
D2=data2[col]
print(D2.head())
````

For each subset, generate a training partition and a test partition using a ratio of 70%−30% 70%−30%

````python
# Se divide la muestra en train y test D1
from sklearn.model_selection import train_test_split
D1train, D1test = train_test_split(D1, test_size=0.3, random_state=3)
print(D1train.shape)
print(D1test.shape)
````

````python
# Se divide la muestra en train y test D2
from sklearn.model_selection import train_test_split
D2train, D2test = train_test_split(D2, test_size=0.3, random_state=3)
print(D2train.shape)
print(D2test.shape)
````

````python
# Se separan las X y Y de los subsets y se les aplica un reshape para poder correr la regresión
XD1train = D1train['RM']
YD1train = D1train['MEDV']
XD1train = XD1train.values.reshape(-1, 1)
YD1train = YD1train.values.reshape(-1, 1)
XD2train = D2train['LSTAT']
YD2train = D2train['MEDV']
XD2train = XD2train.values.reshape(-1, 1)
YD2train = YD2train.values.reshape(-1, 1)
````

Train a linear regression model on both subsets of data:
- Report the mean square error on the test set
- Print the values of ww and w0w0 of the regression equation

````python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(XD1train, YD1train)
lr.coef_, lr.intercept_
````

````python
lr2 = LinearRegression()
lr2.fit(XD2train, YD2train)
lr2.coef_, lr2.intercept_
````

````python
y_pred_d1 = lr.predict(D1test[['RM']])
y_pred_d1
````

````python
y_pred_d2 = lr2.predict(D2test[['LSTAT']])
y_pred_d2
````

````python
from sklearn import metrics
import numpy as np
print('MSE - REGRESSION D1:', metrics.mean_squared_error(D1test['MEDV'], y_pred_d1))
print('MSE - REGRESSION D2:', metrics.mean_squared_error(D2test['MEDV'], y_pred_d2))
````

Generate a graph where you visualize the line obtained by the regression model in conjunction with the training data and the test data

````python
import matplotlib.pyplot as plt
# Plot outputs
plt.scatter(XD1train, YD1train,  color='mediumpurple')
plt.scatter(D1test['RM'], D1test['MEDV'],  color='y')
plt.plot(D1test['RM'], y_pred_d1, color='purple', linewidth=3)

plt.xticks(())
plt.yticks(())
plt.xlabel('RM')
plt.ylabel('MEDV')

plt.show()
````

````python
import matplotlib.pyplot as plt
# Plot outputs
plt.scatter(XD2train, YD2train,  color='mediumpurple')
plt.scatter(D2test['LSTAT'], D2test['MEDV'],  color='y')
plt.plot(D2test['LSTAT'], y_pred_d2, color='purple', linewidth=3)

plt.xticks(())
plt.yticks(())
plt.xlabel('LSTAT')
plt.ylabel('MEDV')

plt.show()
````

- How does the model perform on D1D1 and D2D2? Why?

De acuerdo a los resultados del MSE, se observa que el modelo con D1 es mejor que el estimado con D2. Esto se explica por la forma de la relación de las variables X con la Y de cada subset: mientras que LSTAT (X de D2) disminuye exponencialmente cuando MEDV se incrementa, RM (X de D1) crece linealmente cuando lo hace MEDV. Por lo tanto, una regresión lineal va a predecir mejor en D1 que contiene una X con una relación lineal con la Y, que en D2 donde la X tiene una relación no lineal.
